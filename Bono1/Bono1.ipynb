<span style="color:red">TODO: Create a function to fix the digitation errors for "DIA" column.</span>

```python
days_v = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

def validDay(sday):
    days_s = []
    a = pylev.levenshtein(sday, days_v[0]) 
    index = 0

    for idx in range(1, len(days_v)):
        b = pylev.levenshtein(sday, days_v[idx])    
        if b < a:
            a = b
            index = idx

    return days_v[index]
```

```python
homicides_df["DIA"] = homicides_df["DIA"].apply(validDay)
```
